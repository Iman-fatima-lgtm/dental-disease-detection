```
import gradio as gr
import cv2
import numpy as np
from ultralytics import YOLO  # Assuming you're using YOLOv8

# Load the trained YOLO model
model = YOLO("/content/runs/detect/train2/weights/best.pt")  # Replace "best.pt" with your model path

def predict(image):
    # Run inference on the image
    result = model.predict(image, conf=0.25)
    annotated_img = result[0].plot()  # Get annotated image
    return annotated_img[:, :, ::-1]  # Convert BGR to RGB for display

# Gradio UI
app = gr.Interface(
    fn=predict,
    inputs=gr.Image(type="numpy", label="Upload an image"),
    outputs=gr.Image(type="numpy", label="Prediction"),
    title="Dental Disease Detection",
    description="Upload an  image, and YOLO will predict."
)

# Launch Gradio app
app.launch(share=True)
```
